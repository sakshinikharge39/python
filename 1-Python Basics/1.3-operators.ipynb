```python
#simple calculator 
num1 = float(input("Enter First Number:"))
num2 = float(input("Enter Second Number:"))

add_nums= num1 + num2
sub_nums= num1 - num2
mul_nums= num1 * num2
div_nums= num1 / num2 #quotient
floor_div_nums= num1 // num2
expo_nums= num1 ** num2
modulus_nums = num1 % num2

print(num1)
print(num2)
print("Addition:" , add_nums)
print("Subtraction:" , sub_nums)
print("Multiplication:" , mul_nums)
print("Division:" , div_nums)
print("Floor Divisio:" , floor_div_nums)
print("Exponential:" , expo_nums)
print("Modulus:" , modulus_nums)
```
