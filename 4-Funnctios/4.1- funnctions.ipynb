### Definition:

###### A function is a block of code that performs a specific task. Functions help in organizing code, reusing code, and improving readability.

```python
## syntax
def function_name(parameters):
    """Docstring"""
    # Function body
    return expression
```

```python
# why function
num =24 
if num%2 == 0:
    print("this number is even")
else:
    print("this number is odd ")
```

```python
def even_or_odd(num):
    if num%2 ==0 :
        print("this number is even")
    else:
        print("this number is odd")
        
    
## Call this function
even_or_odd(24)
```

```python
## fuction with multiple parameters 

def add(a,b):
    return a+b 
result= add(2,6)
print(result)
```

```python
## Default Parameters

def greet(name = "guest"):
    print(f"Hello , {name} welcome to hotel paradise")
    
greet("sakshi")
```

```python
### Variable Length Arguments
## Positional And Keywords arguments
def print_numbers(*sakshi):
    for number in sakshi:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"sakshi")
```

```python
## Positional arguments
def print_numbers(*args):
    for number in args:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"sakshi")
```

```python
#keyword arguments
def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(name="sakshi",age= "24")
```

```python
def print_details(*args, **kwargs):
    for val in args:
        print(f"positional arguments:  {val}")
        
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,5, name="sakshi", age="24")
```

```python
### Return statements
def multiply(a,b):
    return a*b

multiply(2,3)
```

```python
### Return multiple parameters
def multiply(a,b):
    return a*b,a

multiply(2,3)
```
